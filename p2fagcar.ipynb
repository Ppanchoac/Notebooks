```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
%matplotlib inline
```

```python
mu=8.53
A=1.2
w=0.2*np.pi
```

```python
def f((x,y), t):
    return [y,mu*(1-x**2)*y-x+A*np.sin(w*t)]
```

```python
f0=[2.0,5.0]
t=np.linspace(0,50,600)
tp=[]
sol=integrate.odeint(f,f0,t)
for i in range(0,600):
    tp.append(t[i])
```

```python
print(sol)
print(len(sol))
```

```python
velocidad=[]
posicion=[]
for i in range(0,600):
    posicion.append(sol[i][0])
    velocidad.append(sol[i][1])
```

```python
plt.plot(tp,posicion,'--',label='Posicion',color='blue')
plt.plot(tp,velocidad,tp,color='yellow',label='Velocidad')
plt.xlabel('Tiempo')
plt.grid(True)
plt.title('Grafico Posicion y Velocidad v/s tiempo')
plt.legend()
```

```python
plt.plot(posicion,velocidad,color='brown')
plt.xlabel('Posicion')
plt.ylabel('Velocidad')
plt.grid(True)
plt.title('Grafico espacio de fase (y(t),v(t))')
plt.savefig('P2-aguilarcarvajal')
```
