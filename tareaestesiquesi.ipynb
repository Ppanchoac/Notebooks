```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.optimize import least_squares
%matplotlib inline
```

```python
y=np.genfromtxt('ltotal')
x=np.genfromtxt('tiempo')
```

```python
def f(x,a,b,c,d):
    return a*np.cos(x*b + c)+d
```

```python
(a,b,c,d),_=curve_fit(f,x,y)
print(a,b,c,d)
```

```python
plt.scatter(x,y)
plt.plot(x,f(x,a,b,c,d),'--',color='red')
```

```python
chicuad=np.sum((y-f(x,a,b,c,d))**2)
```

```python
def ep(a,x,y):
    return a[0]*np.cos(a[1]*x+a[2])+a[3] - y
    #return f(x,a,b,c) = f(x,a[0],a[1],a[2])-y
a_in=[0.5,1,2,1.3,1.8]
```

```python
least_squares(ep,a_in,args=(x,y))
```

```python
a2,b2,c2,d2,_ = least_squares(ep,a_in,args=(x,y)).x
```

```python
plt.scatter(x,y)
plt.plot(x,f(x,a2,b2,c2,d2),label='Least squares',color='black')
plt.plot(x,f(x,a,b,c,d),'--',color='red',label='polyfit')
plt.legend()
```
