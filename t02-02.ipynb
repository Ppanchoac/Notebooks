```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
l1=np.genfromtxt('data1.txt')
x1=[]
for i in range(0,len(l1)):
    x1.append(l1[i][0])
print(x1)
print(len(x1))
y1=[]
for i in range(0,len(l1)):
    y1.append(l1[i][1])
print(y1)
print(len(y1))
```

```python
plt.scatter(x1,y1,color='green')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Datos 1')
```

```python
l2=np.genfromtxt('data2.txt')
x2=[]
for i in range(0,len(l2)):
    x2.append(l2[i][0])
print(x2)
print(len(x2))
y2=[]
for i in range(0,len(l2)):
    y2.append(l2[i][1])
print(y2)
print(len(y2))
```

```python
plt.scatter(x2,y2,color='black')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Datos 2')
```

```python
plt.hist(x1,bins=20)
plt.title('datos 1 con x')
```

```python
plt.hist(y1,bins=20)
plt.title('datos 1 con y')
```

```python
plt.hist(x2,bins=20)
plt.title('datos 2 con x')
```

```python
plt.hist(y2,bins=20)
plt.title('datos 2 con y')
plt.grid(True)
```

```python
promx1=np.sum(x1) / len(x1)
promy1=np.sum(y1) / len(y2)
print('Promedio x1: ',promx1, 'Promedio y1: ', promy1)
promx2=np.sum(x2) / len(x2)
promy2=np.sum(y2) / len(y2)
print('Promedio x2: ',promx2, 'Promedio y2: ', promy2)
```

```python
desvx1= np.sqrt((np.sum(x1-promx1)**2)/len(x1))
desvx2=np.sqrt((np.sum(x2-promx2)**2)/len(x2))
print('Desviacion x1: ',desvx1,'Desviacion x2: ',desvx2)
desvy1= np.sqrt((np.sum(y1-promy1)**2)/len(y1))
desvy2=np.sqrt((np.sum(y2-promy2)**2)/len(y2))
print('Desviacion y1: ',desvy1,'Desviacion y2: ',desvy2)
```
