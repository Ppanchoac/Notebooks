```python
def f(x):
    #f=np.exp(-x**2)
    #f=x**(2.5)
    #f=(1+x**2)**(-1)
    #f=(2+np.cos(x))**(-1)
    f=np.exp(x)*np.cos(4*x)
    #f=np.exp(2-0.5*np.sin(x))

    return f
```

```python
a=float(input('cota inferior: '))
b=float(input('cota superior: '))
n=int(input('numero de trapecios: '))
d=(b-a)/n
x=0.0
z=(f(a)+f(b))*0.5*d
suma=0.0
for i in range(1,n):
    x=a+d*i
    suma=suma + d*f(x)
    print(x,z,suma+z)
```
