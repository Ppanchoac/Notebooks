```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as scs
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
x=np.genfromtxt('manchassolares.txt')
tiempo=[]
manchas=[]
datos=[]
for i in range(0,len(x)):
    if (x[i][4]!=-1):
        tiempo.append(x[i][3])
        manchas.append(x[i][4])
        datos.append([x[i][3], x[i][4]])
```

```python
plt.plot(tiempo,manchas)
plt.xlabel('Tiempo')
plt.ylabel('Numero de manchas')
plt.title('Manchas solares diarias')
plt.grid(True)
plt.savefig('manchasdiariasplot')
```

```python
plt.scatter(tiempo,manchas,s=0.4,color='red',)
plt.xlabel('Tiempo')
plt.ylabel('Numero de manchas')
plt.title('Manchas solares diarias')
plt.grid(True)
plt.savefig('manchasdiarias')
```

```python
(mu, sigma) = scs.norm.fit(manchas)
(n, bins, patches) = plt.hist(manchas, 30, density=1, facecolor='green', alpha=0.75)
y = scs.norm.pdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=2)
```

```python
plt.hist(manchas,bins=30,width=16)
l = plt.plot(bins, y*1333333.33, 'r--', linewidth=2)
plt.title("Numero de manchas")
plt.grid(True)
plt.ylabel("Manchas")
plt.xlabel("Cantidad de dias")
plt.savefig('histogramadias')
```
