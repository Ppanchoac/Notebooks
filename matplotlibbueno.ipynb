```python
x=np.linspace(-10,10,100)
y=np.sin(x)/x
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,'o--')
```

```python
y2=np.exp(-x**2)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.title('el titulo')
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.title('el titulo')
plt.grid(True)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.title('el titulo')
plt.xlim(-6,8)
```

```python
plt.plot(x,y, label='primera')
plt.plot(x,y2, label='segunda')
plt.title('el titulo')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.legend()
```
