Ahora, ajustamos una función de la forma $f(x) = A cos(bx + c) + d$

```python
def f(x,a,b,c,d):
    return (a*np.cos(x*b +c) + d)
```

```python
y=np.genfromtxt('manchasprom.txt')
x=np.genfromtxt('tiempo.txt')
```

A través del paquete de Scipy, curve_fit, haremos un arreglo de la función a través del método de mínimos cuadrados.

```python
(a,b,c,d),_=curve_fit(f,x,y)
print('a=',a,'b=',b,'c=',c,'d=',d)
```

Entonces, $f(x) = 9.310100037066261cos(1.0985037600290664x -163.58100661094338) + 83.01725600364483 )$

```python
plt.plot(x,y)
plt.plot(x,f(x,a,b,c,d),color='red',label='f(x) mediante curve_fit')
plt.grid(True)
plt.title('Ajuste funcional y scatter de datos')
plt.xlabel('Año')
plt.ylabel('Promedio Manchas solares')
plt.legend()
```

```python
plt.plot(x,f(x,a,b,c,d))
plt.title('Ajuste curve_fit')
```

Ahora, haremos el ajuste según least_squares. De esta manera, elegimos los puntos iniciales en a_in similares a los que obtuvimos en curve_fit:

```python
def ep(a,x,y):
    return a[0]*np.cos(a[1]*x+a[2])+a[3] - y
a_in=[9,1.1,-163,80]
```

```python
least_squares(ep,a_in,args=(x,y))
```

```python
a2,b2,c2,d2 = least_squares(ep,a_in,args=(x,y)).x
print('a2=',a2,'b2=',b2,'c2=',c2,'d2=',d2)
```

Luego, $f(x) = -9.31024750280573 cos(1.0985349664044004x -160.49714477791036) + 83.01721233476405$

```python
plt.plot(x,y)
plt.plot(x,f(x,a2,b2,c2,d2),color='red',label='f(x) mediante least_squares')
plt.grid(True)
plt.title('Ajuste funcional y scatter de datos')
plt.xlabel('Año')
plt.ylabel('Promedio Manchas solares')
plt.legend()
```

Calculamos $\chi^2$ para notar qué valor tiene

```python
chicuadrado1=np.sum((y - f(x,a,b,c,d))**2)
chicuadrado2=np.sum((y - f(x,a2,b2,c2,d2))**2)
print(chicuadrado1,chicuadrado2)
```

Notar que $\chi^2$ es menor mediante el arreglo por least_squares. De esta manera, intuimos que es un mejor arreglo.
